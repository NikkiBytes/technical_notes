```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a DiGraph
G = nx.DiGraph()

# Add nodes and edges to the DiGraph
G.add_edge('A', 'B')
G.add_edge('B', 'C')

# Create a MultiDiGraph
MG = nx.MultiDiGraph()

# Add nodes and edges to the MultiDiGraph
MG.add_edge('A', 'B', label='1')
MG.add_edge('A', 'B', label='2')
MG.add_edge('B', 'C', label='1')

# Plotting both graphs
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
nx.draw(G, with_labels=True, font_weight='bold', node_color='skyblue', node_size=1000, arrowsize=20)
plt.title('DiGraph')

plt.subplot(1, 2, 2)
nx.draw(MG, with_labels=True, font_weight='bold', node_color='lightgreen', node_size=1000, arrowsize=20)
plt.title('MultiDiGraph')

plt.tight_layout()
plt.show()
```
